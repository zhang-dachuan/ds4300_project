## Emails Text Analysis

```python
import wordcloud as wc
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import re
from collections import Counter
import pandas as pd
```

```python
# Convert elements to string format
def listToString(s):
    string = ""
    for element in s:
        string += element
    return string
```

```python
def clean_text(text):
    """ docstring"""
    
    # remove leading/trailing whitespaces
    text = text.strip()
    
    # remove punctuation
    text = re.sub('[^\w\s]', '', text)
    text = re.sub('\n', '', text)
    
    #convert to lower case
    text = text.lower()


    return text
```

```python
# filter for words greater than the length of 4
def filter_chars(text):
    list_of_words = []
    
    for word in text.split():
        if len(word) > 4:
            list_of_words.append(word + " ")
        
    
    return list_of_words
```

```python
jeff = pd.read_csv("jeff.csv")

list_of_subjects = jeff["body"]
string_subjects = listToString(list_of_subjects)
subject_cleaned = clean_text(string_subjects)
list_long_subjects = filter_chars(subject_cleaned)
string_long_subjects = listToString(list_long_subjects)
```

```python
# generate word cloud
jeff_cloud = wc.WordCloud().generate(string_long_subjects)
plt.imshow(jeff_cloud)
plt.axis("off")
```
